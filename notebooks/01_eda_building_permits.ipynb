# Edmonton Construction & Housing Activity Intelligence

This project analyzes City of Edmonton building permit activity to identify:
- construction trends over time
- neighbourhood housing growth
- permit seasonality
- development concentration
- construction value patterns

The project uses Python, SQL, SQLite, and dashboard-ready exports built from public City of Edmonton open data.

```python
from pathlib import Path
import pandas as pd

EXPORT_DIR = Path("../data/exports")

yearly = pd.read_csv(EXPORT_DIR / "yearly_trends_completed.csv")
monthly = pd.read_csv(EXPORT_DIR / "monthly_trends_completed.csv")
neighbourhood = pd.read_csv(EXPORT_DIR / "neighbourhood_summary.csv")
permit_type = pd.read_csv(EXPORT_DIR / "permit_type_summary.csv")
seasonality = pd.read_csv(EXPORT_DIR / "seasonality_completed_years.csv")
```

## Key Findings

- Permit activity increased significantly after 2020.
- 2025 had the highest total construction value among completed years.
- Permit activity is strongly seasonal, with peaks during spring and summer months.
- Some neighbourhoods generate many permits but relatively fewer housing units, while others show concentrated high-density development.
- Construction activity appears concentrated in rapidly expanding suburban neighbourhoods.
